#Espaço 3D

O Manim permite criar animações em três dimensões, oferecendo mais realismo e profundidade às representações matemáticas. Para gerar os eixos coordenados, utilizamos o objeto `ThreeDAxes()`.

No espaço 3D, podemos posicionar objetos em diferentes planos e rotacionar a cena para observá-los de vários ângulos. Para controlar a perspectiva da câmera, utilizamos os ângulos $\phi$ e $\theta$:

* $\phi$ (phi): define o ângulo vertical da visão (inclinação da câmera);

* $\theta$ (theta): define o ângulo horizontal (giro ao redor do eixo vertical $z$).


```python
# @title
%%manim -qm plano5

class plano5(ThreeDScene):
    def construct(self):
        Ax3 = ThreeDAxes().add_coordinates()
        self.play(Create(Ax3))
        self.move_camera(phi=65 * DEGREES, theta=60 * DEGREES)
        self.wait(2)


        cone=Cone()
        Toro=Torus()
        esfera=Sphere()
        cilindro=Cylinder()
        cubo=Cube()
        Prisma=Prism()


        cone = Surface(
            lambda u, v: np.array([
                2 * np.cos(v) * np.cos(u),
                2 * np.sin(v) * np.cos(u),
                2 * np.sin(u)]),
            u_range=[-PI, PI],
            v_range=[0, 2 * PI],
            resolution=(15, 32),
            )
        cone.set_fill_by_checkerboard(PURE_GREEN, BLUE,ORANGE,PINK, opacity=0.5)

        paraboloide=Surface(
            lambda u, v: np.array([
                np.cos(v)*u,
                np.sin(v)*u,
                u**2
            ]),v_range=[0, 2 * PI],
            u_range=[0, 1.7],
            resolution=(15, 32) ).scale(0.7)

        paraboloide.set_fill_by_checkerboard(PINK, BLUE, opacity=0.5)

        toro = Surface(
            lambda u, v: np.array([
                (3 + 1 * np.cos(v)) * np.cos(u),
                (3 + 1 * np.cos(v)) * np.sin(u),
                1 * np.sin(v)]),
            v_range=[0, 2 * PI],
            u_range=[-PI, PI],
            resolution=(15, 32)).scale(0.6)

        toro.set_fill_by_checkerboard(RED, BLUE, YELLOW, opacity=0.5)

        catenoide = Surface(
            lambda u, v: np.array([
                0.5 * np.cosh(v / 0.5) * np.cos(u),
                0.5 * np.cosh(v / 0.5) * np.sin(u),
                v]),
            v_range=[-1.2, 1.2],
            u_range=[-PI, PI],
            resolution=(5, 32)).scale(0.7)

        tube2 =Surface(
            lambda u, v: np.array([
                np.sin(v),
                np.cos(u)*np.sin(2*v),
                np.sin(u)*np.sin(2*v)
            ]), u_range=[0, 2*PI], v_range=[0, 2*PI]).scale(1.3)


        tube2.set_fill_by_checkerboard(GREY_BROWN,RED, opacity=0.5)

        self.play(Create(esfera),run_time=7)
        self.wait(2)
```

```python
%%manim -qm RiemannConvergence

class RiemannConvergence(Scene):
    def construct(self):
        axes = Axes(
            x_range=[0, 5, 1],
            y_range=[0, 10, 2],
            x_length=8,
            y_length=5,
            axis_config={"include_tip": True}
        ).add_coordinates()

        labels = axes.get_axis_labels(x_label="x", y_label="f(x)")

        def func(x):
            return 0.5 * x**2 + 1

        graph = axes.plot(func, x_range=[0, 4], color=BLUE)
        graph_label = axes.get_graph_label(graph, label="f(x)", x_val=4, direction=UR)

        area = axes.get_area(graph, x_range=[0, 4], color=BLUE, opacity=0.3)

        self.play(Create(axes), Write(labels))
        self.play(Create(graph), FadeIn(graph_label))
        self.wait(1)

        rect_counts = [3, 6, 12, 24, 50, 100]

        current_rects = axes.get_riemann_rectangles(
            graph, x_range=[0, 4], dx=4/rect_counts[0], input_sample_type="left",
            stroke_width=1, fill_opacity=0.5, color=ORANGE
        )

        count_label = Variable(rect_counts[0], Text("n"), var_type=Integer).to_edge(UP)

        self.play(Create(current_rects), Write(count_label))
        self.wait(1)

        for n in rect_counts[1:]:
            new_rects = axes.get_riemann_rectangles(
                graph, x_range=[0, 4], dx=4/n, input_sample_type="left",
                stroke_width=0.5 if n < 50 else 0, fill_opacity=0.5, color=ORANGE
            )

            self.play(
                Transform(current_rects, new_rects),
                count_label.tracker.animate.set_value(n),
                run_time=1.5
            )
            self.wait(0.5)

        self.play(
            FadeOut(current_rects),
            FadeIn(area),
            FadeOut(count_label)
        )

        integral_tex = MathTex(r"\int_0^4 f(x)dx").to_edge(UP).set_color(BLUE)
        self.play(Write(integral_tex))
        self.wait(2)
```

```python
%%manim -qm FourierLogoDrawing

import numpy as np

class FourierLogoDrawing(Scene):
    def construct(self):
        path = ParametricFunction(
            lambda t: np.array([
                2 * np.cos(t) + 0.5 * np.cos(3*t),
                2 * np.sin(t) - 0.5 * np.sin(3*t),
                0
            ]),
            t_range=[0, TAU], color=BLUE
        ).scale(1.5)


        circles = VGroup()
        vectors = VGroup()

        freqs = [1, -3, 5, -7, 9]
        amplitudes = [1.5, 0.4, 0.2, 0.1, 0.05]

        center = ORIGIN
        time = ValueTracker(0)

        def get_circles_and_vectors():
            group = VGroup()
            curr_center = ORIGIN
            t = time.get_value()

            for f, a in zip(freqs, amplitudes):
                circle = Circle(radius=a, stroke_width=1, stroke_opacity=0.5, color=WHITE)
                circle.move_to(curr_center)

                angle = f * t
                endpoint = curr_center + a * np.array([np.cos(angle), np.sin(angle), 0])
                vector = Line(curr_center, endpoint, buff=0, stroke_width=2, color=YELLOW)

                group.add(circle, vector)
                curr_center = endpoint

            return group

        moving_parts = always_redraw(get_circles_and_vectors)

        dot = Dot(color=ORANGE)
        dot.add_updater(lambda d: d.move_to(moving_parts[-1].get_end()))

        path_trace = TracedPath(dot.get_center, stroke_color=BLUE_A, stroke_width=4)

        self.add(moving_parts, dot, path_trace)

        title = Text("Séries de Fourier", font_size=32).to_edge(UP)
        self.play(Write(title))

        self.play(time.animate.set_value(TAU), run_time=10, rate_func=linear)
        self.wait(2)
```

# Física e Sistemas Complexos

Pêndulo Duplo (Teoria do Caos)

```python
%%manim -qm DoublePendulumChaos

import numpy as np

class DoublePendulumChaos(Scene):
    def construct(self):
        # Parâmetros físicos
        G = 9.8  # Gravidade
        L1 = 2.0  # Comprimento da primeira haste
        L2 = 2.0  # Comprimento da segunda haste
        M1 = 1.0  # Massa 1
        M2 = 1.0  # Massa 2
        DT = 0.02 # Passo de tempo

        def get_derivatives(state):
            theta1, z1, theta2, z2 = state

            delta = theta2 - theta1
            den1 = (M1 + M2) * L1 - M2 * L1 * np.cos(delta)**2

            dz1 = (M2 * L1 * z1**2 * np.sin(delta) * np.cos(delta) +
                   M2 * G * np.sin(theta2) * np.cos(delta) +
                   M2 * L2 * z2**2 * np.sin(delta) -
                   (M1 + M2) * G * np.sin(theta1)) / den1

            den2 = (L2 / L1) * den1

            dz2 = (-M2 * L2 * z2**2 * np.sin(delta) * np.cos(delta) +
                   (M1 + M2) * G * np.sin(theta1) * np.cos(delta) -
                   (M1 + M2) * L1 * z1**2 * np.sin(delta) -
                   (M1 + M2) * G * np.sin(theta2)) / den2

            return np.array([z1, dz1, z2, dz2])

        colors = [RED_A, GREEN_A, BLUE_A]
        initial_thetas = [PI/2, PI/2 + 0.001, PI/2 + 0.002]

        pendulums = VGroup()
        traces = VGroup()
        states = []

        for i in range(3):
            state = np.array([initial_thetas[i], 0.0, PI/2, 0.0])
            states.append(state)

            p = VGroup(
                Line(ORIGIN, DOWN*2),
                Line(DOWN*2, DOWN*4),
                Dot(radius=0.1, color=colors[i])
            ).shift(UP*1)

            trace = TracedPath(p[2].get_center, stroke_color=colors[i], stroke_width=2, dissipating_time=2)

            pendulums.add(p)
            traces.add(trace)

        self.add(traces, pendulums)

        def update_pendulums(mob, dt):
            for i in range(3):
                s = states[i]
                k1 = get_derivatives(s)
                k2 = get_derivatives(s + k1 * DT / 2)
                k3 = get_derivatives(s + k2 * DT / 2)
                k4 = get_derivatives(s + k3 * DT)
                states[i] += (k1 + 2*k2 + 2*k3 + k4) * DT / 6

                t1, _, t2, _ = states[i]

                p1 = np.array([L1 * np.sin(t1), -L1 * np.cos(t1), 0])
                p2 = p1 + np.array([L2 * np.sin(t2), -L2 * np.cos(t2), 0])

                mob[i][0].set_points_as_corners([UP*1, UP*1 + p1])
                mob[i][1].set_points_as_corners([UP*1 + p1, UP*1 + p2])
                mob[i][2].move_to(UP*1 + p2)

        pendulums.add_updater(update_pendulums)

        title = Text("Caos: O Efeito Borboleta", font_size=36, color=WHITE).to_edge(UP)
        self.add(title)

        self.wait(15)
        pendulums.remove_updater(update_pendulums)
        self.wait(1)
```
